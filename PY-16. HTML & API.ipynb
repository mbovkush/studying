```python
def exchange_rates(currency, format='full'):    
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'  
    response = requests.get(url).json()['Valute']    
    data = response[currency]     
    if format == 'full':    
        return data      
    elif format == 'value':    
        return data['Value'] 
```

```python
def currency_name(curr_id):
    url ='https://www.cbr-xml-daily.ru/daily_json.js'  
    response = requests.get(url).json()['Valute']
    for val in response.values():
        if val['ID'] == curr_id:
            return val['Name']
```

Начнем с  формирования запроса к API ВКонтакте методом wall.get():

```python
import requests  
from pprint import pprint  
  
url = 'https://api.vk.com/method/wall.get'  
params = {  
    'domain': 'vk',  
    'filter': 'owner',  
    'count': 10,  
    'offset': 0,  
    'access_token': 'af3e21e1af3e21e1af3e21e13faf47b613aaf3eaf3e21e1ce63bcc88f62276b5ecc252b',  
    'v': 5.95  
}  
response = requests.get(url, params = params)  
response.json()
```

```python
# Видим, что сначала идёт общее количество постов, а по ключу 'items'  —  сами посты. Посмотрим на отдельный пост:
response.json()['response']['items'][0]  
```

```python
# Нужная нам статистика находится в полях 'comments', 'likes' и 'reposts'. 
# Соберем итоговую статистику для каждого поста в словарь stats. 
# В качестве ключа будем использовать начало статьи, в качестве значения — список 
# с тремя интересующими нас метриками и временем публикации: [комментарии, лайки, репосты, дата публикации]
stats = {}  
              
for record in response.json()['response']['items'][:]:  
    title = record['text'][:30]  
    if title:  
        stats[title] = [record['comments']['count'], record['likes']['count'], record['reposts']['count'], record['date'] ]  
pprint(stats)  
```

```python
def get_smm_index(group_name, token):
    url = 'https://api.vk.com/method/wall.get'  
    params = {  
        'domain': group_name,  
        'filter': 'owner',  
        'count': 10,  
        'offset': 0,  
        'access_token': 'af3e21e1af3e21e1af3e21e13faf47b613aaf3eaf3e21e1ce63bcc88f62276b5ecc252b',  
        'v': 5.95  
    }  
    response = requests.get(url, params = params)
    count =0        
    for record in response.json()['response']['items']:
        print(record['comments']['count'])  
        count += record['comments']['count']
        count += record['likes']['count']
        count += record['reposts']['count']
    
    pprint(count)

    url = 'https://api.vk.com/method/groups.getMembers'  
    params = {  
        'group_id':  group_name,
        'v': 5.95,  
        'access_token': 'af3e21e1af3e21e1af3e21e13faf47b613aaf3eaf3e21e1ce63bcc88f62276b5ecc252b' 
    }  
    response = requests.get(url, params = params)  
    data = response.json()  
    all = data['response']['count']
    print(all)
    return count/all


get_smm_index('vk', 'af3e21e1af3e21e1af3e21e13faf47b613aaf3eaf3e21e1ce63bcc88f62276b5ecc252b' )
```
