```python
sample2 = pd.read_csv('sample.csv')
sample2.Age = sample.age.apply(lambda x: x + 1)
display(sample2)
```

```python
sample2.City = sample.city.apply(lambda x: str(x).lower())
display(sample2)
```

```python
def profession_code(s):
    if s == 'Рабочий':
        return 0
    elif s == 'Менеджер':
        return 1
    else:
        return 2
sample2.Profession = sample.profession.apply(profession_code)
display(sample2)
```

```python
def age_category(age):
    if age < 23:
        return 'молодой'
    elif  23 <= age <=35:
        return 'средний'
    else:
        return 'зрелый'

sample['age_category'] = sample.age.apply(age_category)
display(sample)
```

```python
log = pd.read_csv('log.csv', header=None)
log.columns = ['user_id','time','bet','win']
log['user_id'] = log['user_id'].apply(lambda x: x.replace('Запись пользователя № - ',''))

def clean_user_id(s):
    if '#error' in s:
        return ""
    else: 
        return s

log.user_id = log.user_id.apply(clean_user_id)
display(log)
```

```python
# Final task  8.12
# Читаем файл в переменную log
log = pd.read_csv("log.csv",header=None) 
# Переименовываем колонки
log.columns = ['user_id','time','bet','win'] 
# Удаляем строки с ошибками
log = log[~log.user_id.str.contains("#error", na=False)]
# Оставим только значение идентификатора пользователя
log['user_id'] = log['user_id'].apply(lambda x: x.replace('Запись пользователя № - ',''))  
# Удаляем квадратную скобку из поля time
log.time = log.time.str.replace('[','') 
display(log)
```
