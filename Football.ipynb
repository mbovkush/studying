```python
grouped_df = df.groupby(['Position'])['Wage'].sum().sort_values(ascending=False)
print(grouped_df.loc[grouped_df > 5000000])
```

```python
grouped_df = df.groupby(['Position'])[['Wage','Value']].mean().sort_values(['Value'],ascending=False)
print(grouped_df.head(5))
```

```python
wage = df.groupby(['Club'])[['Wage']].agg(['mean', 'median'])
display(wage)
print(wage.loc[wage > 5000000])
```

```python
group = (df.groupby(['Club'])['Wage'].agg(['mean', 'median']))
new_df = (group[group['mean'] == group['median']])
sorted = new_df.sort_values(['mean'], ascending=False)
print(sorted)
display(sorted)
new_df.count()
```
