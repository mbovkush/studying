```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from sklearn.linear_model import LinearRegression
from sklearn import metrics 
data['price_doc'] = data['price_doc'].apply(lambda w: np.log(w + 1))
data = data.drop(['preschool_education_centers_raion', 
                  'kindergarten_km', 'id', 'park_km', 'kremlin_km', 'life_sq'], axis = 1)
data.dropna(inplace=True)
X = data.drop('price_doc', axis = 1)
Y = data['price_doc'].values
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=77)
scaler = RobustScaler()
X_train_transformed = scaler.fit_transform(X_train)
X_test_transformed = scaler.transform(X_test)
reg = LinearRegression().fit(X_train_transformed, Y_train)
y_pred = reg.predict(X_test_transformed)
MSE = np.round(metrics.mean_squared_error(np.exp(Y_test) - 1, np.exp(y_pred) - 1), 0)
print(MSE)
```
